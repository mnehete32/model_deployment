```python
import pandas as pd
```

```python
data = pd.read_csv("../data/census.csv")
```

```python
data["salary"]
```

```python
cat_features = [
"workclass",
"education",
"marital-status",
"occupation",
"relationship",
"race",
"sex",
"native-country",
]
num_features = ['age', 'fnlgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

for cat_feature in cat_features:
    unique_cat = data[cat_feature].unique()
    for category in unique_cat:
        for num_feat in num_features:
            print(cat_feature,num_feat)
            avg_value = data[data[cat_feature] == category][num_feat].mean()
            print(avg_value)
```
